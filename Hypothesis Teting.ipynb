## Hypothesis Testing

```python
import numpy as np
from scipy.stats import t
```

```python
#given data
sample_mean=3050
theoretical_mean=1000+5*600
standard_dev=5*25
sample_size=25
alpha=0.05
```

## Task 1. State the Hypotheses statement:

- H0:The weekly operating cost is consistent with the theoretical model(W=1000+5X)
- H1:The weekly operating cost is higher than the theoretical model.

## TASK-2 Calculate the Test Statistic:

```python
t_stats= (sample_mean-theoretical_mean)/(standard_dev/np.sqrt(sample_size))
t_stats
```

## 3. Determine the Probability and compare:

```python
dof=sample_size-1
critical_value=t.ppf(1-alpha,dof)
critical_value
```

- with t_statistics = -38 which is â‰ˆ 0 so
- p=1
- p > alpha H0 will not be rejected




## TASK-4 Make a Decision:

```python
if t_stats<critical_value:
    print("Accept H0:The operating cost for a franchise in a week is not changed and remained same as $1000+$5X for X=600")
else:
    print("Reject H0:The operating cost for a franchise in a week is changed as greater than $1000+$5X for X=600")
```
