```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1.Load the patient dataset and show summary with info().

```python
#1.Load the patient dataset and show summary with info().
df= pd.read_csv('Patient_Data.csv')
df
```

```python
df.info()
```

## 2.elect only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].

```python
#Selecting  only the columns relevant for billing

billing_column = df[['PatientID', 'Department', 'Doctor', 'BillAmount']]
billing_column.head()
```

## 3.Drop administrative columns like ['ReceptionistID', 'CheckInTime'].

```python
#droping the columns ReceptionistID,CheckInTime
df=df.drop(['ReceptionistID', 'CheckInTime'],axis=1)
df.head()
```

## Use groupby to find total bill amount per department.

```python
 #Group by Department and calculate total bill amount
total_bill_per_department = df.groupby('Department')['BillAmount'].sum().reset_index()

# Rename columns for clarity
total_bill_per_department.columns = ['Department', 'Total_Bill_Amount']

# Display the result
print(total_bill_per_department)
```

## 5.Remove duplicate patient records based on PatientID.

```python
#Removing duplicate patient records based on PatientID.

df=df.drop_duplicates(['PatientID'])
```

## 6.Fill missing BillAmount values with the mean bill amount.

```python
#Fill missing BillAmount values with the mean bill amount
import warnings
warnings.filterwarnings('ignore')
bill=df['BillAmount'].fillna(df['BillAmount'].mean(), inplace=True)
print(bill)
```

## 7.Merge the billing dataset with patient dataset on PatientID.

```python
#7.Merge the billing dataset with patient dataset on PatientID.
patients_df = pd.read_csv("Patient_Data.csv")
billing_df = pd.read_csv("Billing_Data.csv")
merged_df = pd.merge(patients_df, billing_df, on='PatientID', how='inner')
merged_df
```

## 8.Concatenate an additional DataFrame that contains new patients for the current week (row-wise).

```python
#Concatenate an additional DataFrame that contains new patients for the current week
df_old = pd.read_csv("Patient_Data.csv")

# Create a new DataFrame for this week's patients
df_new = pd.DataFrame({
    'PatientID': [201, 202, 203],
    'Name': ['Anita', 'Raj', 'Kiran'],
    'Age': [29, 40, 35],
    'Gender': ['F', 'M', 'M'],
    'Department': ['Cardiology', 'Orthopedics', 'Neurology'],
    'BillAmount': [4500, 6000, 5200]
})

# Concatenate both DataFrames
df_combined = pd.concat([df_old, df_new], ignore_index=True)
df_combined
```

## 9.Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).

```python
#9.Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).
new_cols = pd.DataFrame({
    'InsuranceCovered': [2000, 3000, 2500],
    'FinalAmount': [3000, 4000, 3500]
})

# Concatenate column-wise
df_combined = pd.concat([df, new_cols], axis=1)
df_combined
```
