## Exercise 1: Prime Number

```python
# program for  Prime Number

n = int(input("Enter the number: "))

if n > 1:
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            print(f"The number {n} is not prime")
            break
    else:
        print(f"The number {n} is prime")
else:
    print(f"The number {n} is not prime")   
```

## Exercise 3: Squares of Even/Odd Numbers

```python
# program for square of even/odd
for num in range(100, 201):  # range(100, 201) includes 100 and goes up to 200
    if num % 2 == 0:         # check if the number is even
        print(f"The square of {num} is {num ** 2}")
```

## Exercise 4:Hospital Billing System

```python
# Hospital Billing System

# Given data
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_cost = 300
discount_rate = 0.10  # 10%

# Patient details
days_stayed = 4
num_lab_tests = 3
medicine_charges = 2400

# Calculate individual charges
room_charges = room_charge_per_day * days_stayed
lab_charges = lab_test_cost * num_lab_tests

# Calculate total before discount
total_before_discount = room_charges + doctor_fee + lab_charges + medicine_charges

# Apply discount if applicable
if total_before_discount > 10000:
    discount = total_before_discount * discount_rate
else:
    discount = 0

# Final bill
final_bill = total_before_discount - discount

# Print the bill details
print("----- Hospital Bill Summary -----")
print(f"Room Charges ({days_stayed} days): ₹{room_charges}")
print(f"Doctor Consultation Fee: ₹{doctor_fee}")
print(f"Lab Test Charges ({num_lab_tests} tests): ₹{lab_charges}")
print(f"Medicine Charges: ₹{medicine_charges}")
print(f"Total Before Discount: ₹{total_before_discount}")
print(f"Discount Applied: ₹{discount}")
print(f"Final Bill Amount: ₹{final_bill}")
```
