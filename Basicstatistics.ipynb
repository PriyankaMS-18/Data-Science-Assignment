## Box Plots

```python
#Box plot for Volume Column
import seaborn as sns
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, y='Volume', color='skyblue', width=0.4)
plt.title('Boxplot of Volume column')
plt.ylabel('Volume')
plt.show()
```

-Interquartile Range (IQR):The box represents the middle 50% of data (Q1 to Q3).
-  Approximate values from the plot:
- Q1 (25th percentile) ≈ 3
- Q3 (75th percentile) ≈ 6
- IQR = Q3 - Q1 = 6 - 3 = 3
- Any value greater than Q3 + 1.5 × IQR = 6 + (1.5 × 3) = 10.5 is considered an outlier.
- Most sales volumes are between 3 and 6 units.


```python
#Box plot for Avg Price
plt.figure(figsize=(6, 6))
sns.boxplot(data=df, y='Avg Price', color='skyblue', width=0.4)
plt.title('Boxplot of Avg Price Column')
plt.ylabel('Avg Price')
plt.show()
```

- Third Quartile (Q3): The top of the box is at approximately 10,000.
- First Quartile (Q1): The bottom of the box is very close to 0, perhaps around 1,000.
- Estimated IQR: This makes the interquartile range (Q3 - Q1) approximately 9,000.
- Any value greater than Q3 + 1.5 × IQR = 10,000 + (1.5 × 9000) = 23500 is considered an outlier.
- Outliers: The plot clearly shows three outliers, represented by the individual circles at the top. 
- These points are significantly higher than the rest of the data.

```python
#Box plot for Total Slaes Value column
plt.figure(figsize=(6, 6))
sns.boxplot(data=df, y='Total Sales Value', color='skyblue', width=0.4)
plt.title('Boxplot of Total Sales Value Column')
plt.ylabel('Total Sales Value')
plt.show()
```

- Third Quartile (Q3): The top of the box is at approximately 53,000.
- First Quartile (Q1): The bottom of the box is very low, near 0, perhaps around 3,000.
- Estimated IQR: This makes the interquartile range (Q3 - Q1) approximately 51,000

```python
#Box plot for Total Discount Rate (%) column
plt.figure(figsize=(6, 6))
sns.boxplot(data=df, y='Discount Rate (%)', color='skyblue', width=0.4)
plt.title('Boxplot of Discount Rate (%) Column')
plt.ylabel('Discount Rate (%)')
plt.show()
```

## Bar Chart Analysis for Categorical Column

```python
# Bar Chart Analysis for Categorical Column:
Categorical_column = df.select_dtypes(include=['object']).columns
Categorical_column
```

```python
# Count the occurrences of each date
date_counts = df['Date'].value_counts().sort_index()

# Plot bar chart
plt.figure(figsize=(10, 5))
plt.bar(date_counts.index, date_counts.values, color='purple', edgecolor='black')
plt.title('Bar Chart of  Date')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
