```python
#printing the grade according to the marks obtained

student_marks=int(input("enter the marks"))

if student_marks>=90:
    print("Grade:A")
elif student_marks>=75:
    print("Grade:B")
elif student_marks>=50:
    print("Grade:C")
elif student_marks<50:
    print("Fail!!")
```

## 11.Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'.
Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows

```python
#creating the dictionary manualy
dict={
    'Name':['Ram','Priya','Jhon','Ram','Jack','Tina','Amith','Riya','Jhon','Meena'],
    'Age': [25, 28, 22, 25, 30, 25, 29, 22, 24, 27],
    'City': ['Delhi', 'Mumbai', 'Chennai', 'Delhi', 'Delhi', 'Delhi', 'Bangalore', 'Chennai', 'Pune', 'Mumbai'],
    'Salary': [45000, 52000, np.nan, 45000, 50000, 45000, 53000, 48000, np.nan, 51000]
}
dict
```

```python
import pandas as pd
import numpy as np

df=pd.DataFrame(dict) #converting th dict into Data Frame
df
```

```python
#applying info()
df.info()
```

```python
# applying describe()
df.describe(include='all')
```

```python
#selecting name and city columns

print(df[['Name','City']])
```

```python
#droping column city

df.drop(columns=['City'])
```

```python
# Fill any missing values in 'Salary' column with the mean

df['Salary'].fillna(df['Salary'].mean(), inplace=True)
df
```

```python
#Removing any duplicate rows
df=df.drop_duplicates()
df = df.reset_index(drop=True)
df
```
